```python
import pickle, os
import pandas as pd
```

```python
path = "residues_embs"
```

# Data

```python
with open("../training_data/5.Embeddings/embeddings.pkl", "rb") as f:
    embs = pd.concat(pickle.load(f).values())

embs
```

```python
with open("../training_data/6.Training_sets/residues_train.pkl", "rb") as f:
    train_features = pickle.load(f)

train_features
```

```python
train_data = train_features[["Residues", "Label"]].merge(
    embs,
    on=[("Residues", c) for c in train_features["Residues"].columns] + [("Label", "label"),],
    indicator=True
)

assert all(train_data["_merge"] == "both") and len(train_data) == len(train_features)

train_data = train_data.drop(columns="_merge")

train_data
```

```python
with open("../training_data/6.Training_sets/residues_test.pkl", "rb") as f:
    test_features = pickle.load(f)

test_features
```

```python
test_data = test_features[["Residues", "Label"]].merge(
    embs,
    on=[("Residues", c) for c in test_features["Residues"].columns] + [("Label", "label"),],
    indicator=True
)

assert all(test_data["_merge"] == "both") and len(test_data) == len(test_features)

test_data = test_data.drop(columns="_merge")

test_data
```

```python
def process_dataframe(df):
    df.index = df["Residues"][["pdb","label_asym_id", "label_seq_id"]].apply(lambda x: "_".join(x), axis=1)
    df = df.drop(columns=["Residues"], level=0)
    df.columns = map(lambda x: "_".join(map(str, x)), df.columns.values)
    df.loc[:,'Label_label'] = df['Label_label'].astype("category")
    return df
```

```python
train = process_dataframe(train_data)

train
```

```python
test = process_dataframe(test_data)

test
```

# Training

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
traindataset = TabularDataset(train)

traindataset
```

```python
traindataset.columns[ traindataset.isna().any() ]
```

```python
testdataset = TabularDataset(test)
```

```python
from sklearn.metrics import matthews_corrcoef
```

```python
from autogluon.core.metrics import make_scorer

mcc = make_scorer(
    name='mcc',
    score_func=matthews_corrcoef,
    optimum=1,
    greater_is_better=True
)
```

```python
print(help(TabularPredictor))
```

```python
print(help(TabularPredictor.fit))
```

```python
if not os.path.isdir(path):
    predictor = TabularPredictor(
        label='Label_label', problem_type="binary", eval_metric=mcc,
        path=path
    )
    
    fit = predictor.fit(
        traindataset,
        presets="best_quality",
        time_limit=3600,
        num_gpus=0, num_cpus=30,
        verbosity=2,
        auto_stack=False, num_stack_levels=0,
        fit_weighted_ensemble=False
    )
else:
    fit = predictor.load(path)
```

```python
leaderboard = fit.leaderboard(testdataset)
leaderboard.iloc[:40]
```

```python
fit.model_best
```

```python
leaderboard.query(f"model == '{fit.model_best}'")["score_val"].item()
```

```python
# fit.set_model_best("NeuralNetFastAI_r37")
```

# Results

```python
trainresults = fit.evaluate(traindataset, detailed_report=True)
trainresults
```

```python
results = fit.evaluate(testdataset, detailed_report=True)
results
```

```python
print("mcc", results["mcc"])
print("macro-f1", results["classification_report"]["macro avg"]["f1-score"])
print(results["confusion_matrix"])
```

```python
with open("../training_data/7.Extra_set/embeddings.pkl", "rb") as f:
    extra = process_dataframe(pd.concat((
        pickle.load(f).values()
    )))

extra
```
