# Split

Split PDBs stratifying pocket labels, 80/20, and check afterwards the splitting of single-residue labels.

```python
from sklearn.model_selection import StratifiedGroupKFold
```

```python
splitter = StratifiedGroupKFold(n_splits=int(1/0.2), shuffle=True, random_state=0)

split = tuple(splitter.split(
    X = tuple(range(len(pockets))), 
    y = pockets[("Label", "label")],
    groups = pockets[("Pockets", "pdb")]
))[0]

split
```

```python
train = pockets.iloc[split[0]]
train
```

```python
test = pockets.iloc[split[1]]
test
```

```python
print("Train PDBs, Test PDBs, % split")
len(train[("Pockets", "pdb")].unique()), len(test[("Pockets", "pdb")].unique()), len(test) / ( len(train) + len(test) )
```

## Pockets

### Train set outliers

```python
print("PDBs, total and positive training pockets before outlier removal:", len(train[("Pockets", "pdb")].unique()), len(train), train[("Label", "label")].sum())

train = train[
    (
        ( train[("Pockets", "nres")] - train[("Pockets", "nres")].mean() ).abs() / train[("Pockets", "nres")].std()
    ) < 3
]

print("PDBs, total and positive training pockets after:", len(train[("Pockets", "pdb")].unique()), len(train), train[("Label", "label")].sum())
```

### Balance

```python
print("Train pockets, Neg. pockets, Pos. pockets, % pos., ratio pos/neg")

len(train), \
len(train.loc[train[("Label", "label")] == 0]), \
len(train.loc[train[("Label", "label")] == 1]), \
len(train.loc[train[("Label", "label")] == 1]) / len(train), \
len(train.loc[train[("Label", "label")] == 1]) / len(train.loc[train[("Label", "label")] == 0])
```

```python
print("Test pockets, Neg. pockets, Pos. pockets, % pos., ratio pos/neg")

len(test), \
len(test.loc[test[("Label", "label")] == 0]), \
len(test.loc[test[("Label", "label")] == 1]), \
len(test.loc[test[("Label", "label")] == 1]) / len(test), \
len(test.loc[test[("Label", "label")] == 1]) / len(test.loc[test[("Label", "label")] == 0])
```

## Residues

```python
from tqdm.notebook import tqdm
import pandas as pd
```

```python
train_residues = pd.DataFrame({
    pdb: {
        "len": len(feats),
        "pos": feats[("Label", "label")].sum()
    }
    for pdb, feats in features.items()
    if pdb in train[("Pockets", "pdb")].unique()
}).T

train_residues
```

```python
test_residues = pd.DataFrame({
    pdb: {
        "len": len(feats),
        "pos": feats[("Label", "label")].sum()
    }
    for pdb, feats in features.items()
    if pdb in test[("Pockets", "pdb")].unique()
}).T

test_residues
```

```python
print("% split of test acc. to # of residues")
test_residues.len.sum() / ( train_residues.len.sum() + test_residues.len.sum() )
```

```python
print("Train residues, Neg. residues, Pos. residues, % pos., ratio pos/neg")

train_residues.len.sum(), train_residues.len.sum() - train_residues.pos.sum(), train_residues.pos.sum(), train_residues.pos.sum() / train_residues.len.sum(), train_residues.pos.sum() / ( train_residues.len.sum() - train_residues.pos.sum() )
```

```python
print("Test residues, Neg. residues, Pos. residues, % pos., ratio pos/neg")

test_residues.len.sum(), test_residues.len.sum() - test_residues.pos.sum(), test_residues.pos.sum(), test_residues.pos.sum() / test_residues.len.sum(), test_residues.pos.sum() / ( test_residues.len.sum() - test_residues.pos.sum() )
```

#### Imputation

```python
train_residues_imputed = pd.concat((
    train_residues[("Residues", "pdb")], # add back the pdb column because .groupby drops it
    ( # pdb-wise, column-mean imputation
        train_residues
        .groupby(("Residues", "pdb"), sort=False)
        .transform(lambda x: ( # column-wise function
            x.fillna(x.mean()) 
            if x.isna().any() and x.dtype.kind in 'biufc' 
            else x
        ))
    )),
    axis=1
)

train_residues_imputed
```

```python
train_residues_imputed.loc[ train_residues_imputed.isna().any(axis=1) ].value_counts(("Residues", "pdb"))
```

```python
test_residues = pd.concat(
    feats
    for pdb, feats in features.items()
        if pdb in test[("Pockets", "pdb")].unique()
)

test_residues
```

```python
test_residues_imputed = pd.concat((
    test_residues[("Residues", "pdb")], # add back the pdb column because .groupby drops it
    ( # pdb-wise, column-mean imputation
        test_residues
        .groupby(("Residues", "pdb"), sort=False)
        .transform(lambda x: ( # column-wise function
            x.fillna(x.mean()) 
            if x.isna().any() and x.dtype.kind in 'biufc' 
            else x
        ))
    )),
    axis=1
)

test_residues_imputed
```

```python
train_residues_imputed.to_pickle(f"{sets_path}/residues_train.pkl")
test_residues_imputed.to_pickle(f"{sets_path}/residues_test.pkl")
```
